**import libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , precision_score, recall_score, accuracy_score
```

```python
from sklearn.ensemble import RandomForestClassifier
import os
os.listdir()
```

**Load Data**

```python
import zipfile

zip_path = "tesla-directional-move-prediction-from-technical-indicators (1).zip"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("tesla_data")   # folder name you want
```

```python
os.listdir("tesla_data")
```

```python
train = pd.read_csv("tesla_data/train.csv")
test = pd.read_csv("tesla_data/test.csv")

print(train.shape)
print(test.shape)
```

```python
x = train.drop(["y"],axis = 1)
```

```python
y = train["y"]
```

```python
y
```

**Train Test Split**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

**Random Forest**

```python
model = RandomForestClassifier(n_estimators=1225,
    max_depth=12,
    min_samples_split=20,
    min_samples_leaf=5,
    max_features=0.5)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

**Accuracy**

```python
acc = print("accuracy",accuracy_score(y_test, y_pred))
```
